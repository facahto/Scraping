```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/Category:American_women_lawyers'
req = requests.get(url)
```

```python
soup = BeautifulSoup(req.text,'lxml') #html script
main = soup.find('div', class_="mw-category mw-category-columns").findAll('li')
for m in main:
  print(m)
```
